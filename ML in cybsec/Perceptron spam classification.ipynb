```python
import string
exclude = set(string.punctuation)
nopunct_lower = lambda x: ''.join(ch for ch in x if ch not in exclude).lower()
```

```python
nopunct_lower('bkBV<><>/lvf///...YTVKDBLnjhkvgdcsctYV?,,,,,.')
```

```python
spam_words = [nopunct_lower(x) for x in spam_words]
```

```python
spam_words
```

```python
from collections import Counter

def most_frequent(words, rank):
    occurence_count = Counter(words)
    return occurence_count.most_common(rank)
```

```python
most_frequent(spam_words, 30)
```

y = 3*free + 4*prize

```python
sample = [
    [1, 1, 0, 3*1 + 4*0, 'ham'],
    [2, 0, 1, 3*0 + 4*1, 'ham'],
    [2, 1, 1, 3*2 + 4*1, 'spam']
]

df_test = pd.DataFrame(sample, columns = ['sms_no', 'free', 'prize', '3free + 4prize', 'spam_or_ham'])
```

```python
df_test
```

```python
def calculate_word_amount(sentence, target):
    #разбиваем предложение на массив слов
    sentence = sentence.split()
    sentence = [nopunct_lower(x) for x in sentence]
    return sum(1 for x in sentence if x == target)
```

```python
calculate_word_amount('I want a free prize tomorroe Free#', 'free')
```

```python
df['free_count'] = df['sms'].apply(lambda x: calculate_word_amount(x, 'free'))
df['prize_count'] = df['sms'].apply(lambda x: calculate_word_amount(x, 'prize'))
df['3f+4p'] = 3*df['free_count'] + 4*df['prize_count']
```

```python
df.head()
```

пусть при значении формулы <=4 - не спам

```python
df['f_p_decision'] = df['3f+4p'].apply(lambda x: 'ham' if x<=4 else 'spam')
df['f_p_decision_true'] = df['f_p_decision'] == df['spam_or_ham']
```

```python
df.head(10)
```

```python
df['f_p_decision_true'].value_counts()
```

```python
df['spam_or_ham_num'] = df['spam_or_ham']. apply(lambda x: 1 if x =='spam' else 0)
```

```python
df.head()
```

```python
X = df[['free_count', 'prize_count']]
y = df['spam_or_ham_num']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=0)
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron(max_iter=80, eta0=0.1, random_state=0)
p.fit(X_train, y_train)
```

```python
y_pred = p.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
print('количество неправильно предсказанных значений: %d' % (y_test!=y_pred).sum())
print('accuracy: %.2f' % accuracy_score(y_test, y_pred))
```

## спам-фисльтр с svm

```python
from sklearn.svm import SVC
svm = SVC(kernel='linear', C=1, random_state=0)
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
```

```python
print('количество неправильно предсказанных значений: %d' % (y_test!=y_pred).sum())
print('accuracy: %.2f' % accuracy_score(y_test, y_pred))
```
